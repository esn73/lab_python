```python
# total_bill(영수증 금액)이 최댓값인 행을 출력
tips[tips['total_bill']==tips['total_bill'].max()]
```

```python
# total_bill이 최솟값인 행을 출력
tips[tips['total_bill']==tips['total_bill'].min()]
```

```python
# total_bill이 최댓값이거나 최솟값인 행을 출력
tips[(tips['total_bill'] == tips['total_bill'].max()) | 
     (tips['total_bill'] == tips['total_bill'].min())]
```
